```python
import pandas as pd
import requests 
from datetime import datetime
from pathlib import *
import numpy as np

base_file = Path('/home/tomi/Pandas')
file_path_crib = '{category}/{year}-{month:02d}/{category}-{day:02d}-{month:02d}-{year}.csv'
```

```python
categories =[
  
    {
        'name': 'museos',
        'url': 'https://datos.cultura.gob.ar/dataset/37305de4-3cce-4d4b-9d9a-fec3ca61d09f/resource/4207def0-2ff7-41d5-9095-d42ae8207a5d/download/museos_datosabiertos.csv'
    },
    {
        'name': 'cines',
        'url': 'https://datos.cultura.gob.ar/dataset/37305de4-3cce-4d4b-9d9a-fec3ca61d09f/resource/392ce1a8-ef11-4776-b280-6f1c7fae16ae/download/cine.csv'

    },
    {
        'name': 'biliotecas_populares',
        'url': 'https://datos.cultura.gob.ar/dataset/37305de4-3cce-4d4b-9d9a-fec3ca61d09f/resource/01c6c048-dbeb-44e0-8efa-6944f73715d7/download/biblioteca_popular.csv'
    }
]
```

## Museos

```python
category_dict = categories[0]
r = requests.get(category_dict['url'])
r.encoding = 'utf-8'
```

```python
now = datetime.now()

file_path = file_path_crib.format(category=category_dict['name'], year=now.year, month=now.month, day=now.day)

m_path = base_file / file_path
m_path.parent.mkdir(parents=True, exist_ok=True)

# Guardo archivo
with open(m_path, 'w') as f_out:
    f_out.write(r.text)
```

## Transform

```python
df = pd.read_csv(m_path)
```

```python
df.head()
```

```python
renamed_columns_museos = {
    'Cod_Loc': 'cod_localidad',
    'IdProvincia': 'id_provincia',
    'IdDepartamento': 'id_departamento',
    'direccion': 'domicilio',
    'CP': 'codigo_postal',
    'telefono': 'numero_telefono',
    'Mail': 'mail',
    'Web': 'web',
}

df = df.rename(columns = renamed_columns_museos)

campos = ['cod_localidad','id_provincia','id_departamento','categoria','provincia','localidad','nombre','domicilio','codigo_postal','numero_telefono','mail','web']

df_museos = df[campos]
```
